Q1.
정수 배열 arr가 주어집니다. arr의 각 원소에 대해 값이 50보다 크거나 같은 짝수라면 2로 나누고, 50보다 작은 홀수라면 2를 곱하고 다시 1을 더합니다.

이러한 작업을 x번 반복한 결과인 배열을 arr(x)라고 표현했을 때, arr(x) = arr(x + 1)인 x가 항상 존재합니다. 이러한 x 중 가장 작은 값을 출력하시오.

단, 두 배열에 대한 "="는 두 배열의 크기가 서로 같으며, 같은 인덱스의 원소가 각각 서로 같음을 의미합니다.

image.png

https://school.programmers.co.kr/learn/courses/30/lessons/181881

```python
arr = [1, 2, 3, 100, 99, 98]

i = 0
while True:
    i += 1

    tmp = arr[:]
    for j in range(len(arr)):
        if arr[j] >= 50 and arr[j] % 2 == 0:
           arr[j] /= 2
        elif arr[j] < 50 and arr[j] % 2 != 0:
           arr[j] = (arr[j] * 2) + 1

    if tmp == arr:
        break

print(i - 1)
```

Q2.
10진수 숫자 n을 입력받았을 때, 이를 이진수로 표현하는 배열 binary를 출력하시오.

```python
n = int(input("10진수를 입력하세요: "))
binary = []

while n > 0:
    remainder = n % 2
    binary.append(remainder)
    n //= 2

if not binary:
    binary = [0]

binary.reverse()
print(binary)
```
